# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading the file 
df=pd.read_csv('./zomato.csv')
df.head()
```

```python
df.shape   # to check number of rows n columns
```

```python
df.columns    # to check the column names 
```

```python
df.isnull().sum()   # finding out if there are any null values 
```

```python
df= df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis =1)# dropping colums that are not necessary
df.head(5)
```

```python
df.info()
```

```python
df.columns
```

### Dropping duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
# finding all Unique values in columns
for i in df.columns:
    print(df[i].unique())
```

### Handling Rate Column

```python
df['rate'].unique()
```

```python
def rate_(value):
    if value=='NEW' or value=='-':
        return np.nan
    else:
        value= str(value).split('/')
        value=value[0]
        return float(value)
    
```

```python
df['rate']= df['rate'].apply(rate_)
```

```python
df['rate'].head()
```

```python
df.isnull().sum()
```

```python
# Handling null values n replacing it with mean

df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
# dropping null values  in other columns

df.dropna(inplace=True)
df.shape
```

```python
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'costfor2', 'listed_in(type)':'Type'}, inplace=True)
df.head(3)
```

```python
# we are dropping listed in city because we have location (and almost both are same)
df=df.drop(['listed_in(city)'],axis=1)   # axis=1 is for columns
df
```

### Handling Costfor2 Column

```python
df['costfor2'].unique()
```

```python
def comma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
```

```python
df['costfor2']=df['costfor2'].apply(comma)
df['costfor2'].unique()
```

### Handling restaurent type column

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type= df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_typelessthan1000= rest_type[rest_type<1000]
rest_typelessthan1000
```

```python
# clustering all these restuarents in one group called others whose values are less than 1000

def resttypename(value):
    if (value in rest_typelessthan1000):
        return 'others'
    else:
        return value
```

```python
df['rest_type']=df['rest_type'].apply(resttypename)
df['rest_type'].value_counts()
```

```python
df.shape
```

```python
df.head()
```

### Handling location column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending=False)
location.unique()
```

```python
location200= location[location<200]
def loca(value):
    if (value in location200):
        return 'others'
    else:
        return value        
```

```python
df['location']=df['location'].apply(loca)
df['location'].value_counts()
```

### Handling Cuisines column

```python
df['cuisines'].value_counts(ascending=False)
```

```python
df['cuisines'].unique()
```

```python
cuisine=df['cuisines'].value_counts(ascending=False)
cuisine.unique()
```

```python
cuisine_lessthan90= cuisine[cuisine<90]
cuisine_lessthan90
```

```python
def cui(value):
    if value in cuisine_lessthan90:
        return 'others'
    else:
        return value
```

```python
df['cuisines']=df['cuisines'].apply(cui)
df['cuisines'].value_counts()
```

```python
df.head()   
```
